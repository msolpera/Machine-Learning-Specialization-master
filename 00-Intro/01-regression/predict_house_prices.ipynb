```python
import pandas as pd
```

```python
sales = pd.read_csv('home_data.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sqft_living', y='price', data=sales.loc[1:5000])
plt.show()
```

```python
# Usando sklearn train_test_split
from sklearn.model_selection import train_test_split

training_set, test_set = train_test_split(sales, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

# Preparar las caracterÃ­sticas y el objetivo
X = training_set[['sqft_living']]
y = training_set['price']
```

```python
sqft_model = LinearRegression()
```

```python
sqft_model.fit(X, y)
```

```python
# Predecir en el conjunto de prueba
X_test = test_set[['sqft_living']]
y_test = test_set['price']
y_pred = sqft_model.predict(X_test)
```

```python
# Calcular mÃ©tricas
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
print(f"R-squared: {r2}")
```

```python
# Explorar coeficientes y tÃ©rmino independiente
coef = sqft_model.coef_[0]
intercept = sqft_model.intercept_

# Imprimir detalles del modelo
print(f"Coeficiente (pendiente): {coef}")
print(f"TÃ©rmino independiente (intercepto): {intercept}")
```

```python
advanced_features = [
'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode',
'condition', # condition of house				
'grade', # measure of quality of construction				
'waterfront', # waterfront property				
'view', # type of view				
'sqft_above', # square feet above ground				
'sqft_basement', # square feet in basement				
'yr_built', # the year built				
'yr_renovated', # the year renovated				
'lat', 'long', # the lat-long of the parcel				
'sqft_living15', # average sq.ft. of 15 nearest neighbors 				
'sqft_lot15', # average lot size of 15 nearest neighbors 
]
```

```python
training_set_af, test_set_af = train_test_split(sales, test_size=0.2, random_state=0)
```

```python
X_af = training_set_af[advanced_features]
y_af = training_set_af['price']
```

```python
sqft_model_af = LinearRegression()
```

```python
sqft_model_af.fit(X_af, y_af)
```

```python
X_test_af = test_set_af[advanced_features]
y_test_af = test_set_af['price']
y_pred_af = sqft_model_af.predict(X_test_af)
```

```python
from sklearn.metrics import max_error
mse = mean_squared_error(y_test_af, y_pred_af)
rmse = np.sqrt(mse)
r2 = r2_score(y_test_af, y_pred_af)
max_e = max_error(y_test_af, y_pred_af)
print(f"Max error: {max_e}")
print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
print(f"R-squared: {r2}")
```
